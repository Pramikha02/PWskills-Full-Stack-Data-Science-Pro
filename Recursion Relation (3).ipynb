# 2. Given a recurrence relation, solve it using the substitition method:
a. T(n) = T(n-1) + c
b. T(n) = 2T(n/2) + n
c. T(n) = 2T(n/2) + c
d. T(n) = T(n/2) + c

Let's solve each recurrence relation using the substitution method:

a. T(n) = T(n-1) + c

Let's substitute (n-1) for n repeatedly until we reach the base case:


T(n) = T(n-1) + c 
     = T(n-2) + 2c 
     = T(n-3) + 3c
     = ......

T(n) = T(1) + (n-1)c


b. T(n) = 2T(n/2) + n

Let n = 2^k. Substitute n/2 for n repeatedly:


T(2^k) = 2T(2^{k-1}) + 2^k 
       = 2^2T(2^{k-2}) + 2^k 
       = ........
       = 2^kT(1) + k2^k 
       = nT(1) + nlog_2(n)

T(n)   = nT(1) + nlog_2(n)


c. T(n) = 2T(n/2) + c

Using the same substitution method:

T(2^k) = 2T(2^{k-1}) + c 
       = 2^2T(2^{k-2}) + 2c 
       = ......
       = 2^kT(1) + kc
       = nT(1) + cn

So, T(n) = nT(1) + cn


d. T(n) = T(n/2) + c

Again, using the substitution method:

T(2^k) = T(2^{k-1}) + c 
       = T(2^{k-2}) + c + c
       = .......
       = T(1) + kc 
       = T(1) + log_2(n). c

       

These are the solutions for each recurrence relation using the substitution method.


# 3. Given a recurrence relation, solve it using the recursive tree approach:
a. T(n) = 2T(n-1) +1
b. T(n) = 2T(n/2) + n

````python
Let's solve each recurrence relation using the recursive tree approach:

a. \(T(n) = 2T(n-1) + 1\)

We can represent the recursive tree for this recurrence relation as follows:

```
             T(n)
            /   \
       T(n-1)   T(n-1)
         / \      / \
    T(n-2) T(n-2) T(n-2) T(n-2)
      / \   / \   / \   / \
   ...  ... ... ... ... ...
```

Each level of the tree represents the recursive calls, and at each level, the size of the problem decreases by 1. The work done at each level is constant (1).

The total work done can be calculated by summing the number of nodes at each level of the tree:

T(n) = 2^n.T(1) + 2^{n-1}.1 + 2^{n-2}.1 + ..... + 2^2 . 1 + 2^1. 1 + 2^0.1


Simplifying the expression, we get:

T(n) = 2^n . T(1) + 2^n - 1

So, T(n) = O(2^n)



b. T(n) = 2T(n/2) + n

The recursive tree for this recurrence relation looks like:

```
            T(n)
           /   \
     T(n/2)     T(n/2)
     /    \     /    \
T(n/4) T(n/4) T(n/4) T(n/4)
  / \    / \   / \    / \
... ... ... ... ... ... ...
```

At each level, the size of the problem is halved (since (n/2) is passed to the recursive calls, and the work done is linear proportional to n).

The total work done is the sum of the work done at each level:


T(n) = 2^0.n + 2^1.(n/2) + 2^2.(n/4) + ......... + 2^{log_2 (n - 1)}.1


This is a geometric series, and the sum is O(nlogn)

T(n) = O(nlogn)
````
